```python
import cv2,time
video=cv2.VideoCapture(0)
a=1
while True:
    a=a+1
    check,frame=video.read()
    print(frame)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Capture",gray)
    key=cv2.waitKey(1)
    if key==ord('q'):
        break
        
print(a)
video.release()
cv2.destroyAllWindows()
    
```

```python
#Face detection through saved image
```

```python
face_cascade=cv2.CascadeClassifier(r"C:\Users\Sony\Desktop\haarcascade_frontalface_default.xml")
```

```python
img=cv2.imread("C:\\Users\\Sony\\Desktop\\modi.jpg")
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
faces=face_cascade.detectMultiScale(gray,1.1,4)
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    
```

```python
cv2.imshow('img',img)
cv2.waitKey()
```
